# 02 â€” HAG processing\n\nCorre el pipeline de HAG y valida que tenga `HeightAboveGround`.

```python

import subprocess, json
from pathlib import Path

ROOT = Path(".").resolve()
DATA = ROOT / "data"
RAW  = DATA / "raw" / "sample.laz"
PROC = DATA / "processed"
PIPES = ROOT / "pipelines"
LAS_HAG = PROC / "sample_hag.las"
PIPE_MAKE_HAG = PIPES / "pipeline_make_hag.json"

for d in (PROC, PIPES): d.mkdir(parents=True, exist_ok=True)

assert RAW.exists(), f"Falta RAW: {RAW}. CorrÃ© 00_prepare_raw.ipynb"
assert PIPE_MAKE_HAG.exists(), f"Falta pipeline: {PIPE_MAKE_HAG}. CorrÃ© 01_pipelines_setup.ipynb"

def run(cmd):
    print("$", " ".join(cmd))
    p = subprocess.run(cmd, capture_output=True, text=True)
    if p.stdout: print(p.stdout)
    if p.returncode != 0:
        print(p.stderr); raise RuntimeError("Fallo de comando")
    return p.stdout

run(["pdal", "pipeline", str(PIPE_MAKE_HAG)])
out = run(["pdal", "info", str(LAS_HAG), "--all"])
print("Â¿Incluye HeightAboveGround?", "HeightAboveGround" in out)
```
